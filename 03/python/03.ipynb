```python
def is_symbol(c):
    return not (c in ["."] + [str(x) for x in range(10)])

is_symbol("a")
```

```python
def is_symbol_at(index=3, line="..35..633."):
    if len(line) == 0:
        return False
    if index < 0:
        return False
    if index >= len(line):
        return False

    return is_symbol(line[index])
```

```python
def find_symbol_around(
    index=5, line="467..114..", line_above="", line_below="...*......"
):
    look_up = [i + index for i in [-1, 0, +1]]

    for i in look_up:
        contains_symbol = (
            is_symbol_at(i, line)
            or is_symbol_at(i, line_above)
            or is_symbol_at(i, line_below)
        )

        if contains_symbol:
            return True
    return False


find_symbol_around()
```

```python
def extract_number(
    line="313.....753.....................596............*................270..../........*........................38.......836..850..914......942*215",
    line_above="..............+...994..........#.........*..244.457.....*...........867.........829.....469.....#...........................*...............",
    line_below="........*............10.525.515.......417$........976...........*..............878.613/.......*247.......*........+.........................",
):
    line_sum = 0
    current_num = ""
    is_symbol_around = False
    for index, c in enumerate(line):
        if c not in [f"{i}" for i in range(10)]:
            if is_symbol_around:
                is_symbol_around = False
                line_sum += int(current_num)
            current_num = ""

        else:
            current_num = current_num + c

            is_symbol_around = is_symbol_around or find_symbol_around(
                index, line, line_above, line_below
            )

    if is_symbol_around:
        line_sum += int(current_num)
    return line_sum
```

```python
def check_valid(data=data, pos=(1,3)):
    x,y = pos

    check_pos = []
    for xx in range(max(0, x-1), min(x+2, len(data))):
        line = []
        for yy in range(max(0, y-1), min(y+2, len(data[x]))):
            line.append((xx,yy))
        check_pos.append(line)

    total_nums_around = 0
    for line in check_pos:
        if(line == []): continue

        chars = []
        for pos in line:
            c = data[pos[0]][pos[1]]
            chars.append(c)
        is_nums = ["1" if c.isnumeric() else "0" for c in chars]
        to_check = ''.join(is_nums)

        map ={
            "111" : 1,
            "110" : 1,
            "011" : 1,
            "101" : 2,
            "100" : 1,
            "010" : 1,
            "001" : 1,
            "000" : 0
        }
        
        num_of_nums_in_line = map[to_check]
        total_nums_around += num_of_nums_in_line
    return total_nums_around == 2

check_valid()
```

```python
def find_gear_positions(line_index=3, line="...*......"):
    found = []
    for i, c in enumerate(line):
        if(c == "*"):
            if check_valid(data, (line_index, i)):
                found.extend([(line_index, i)])
    return found
find_gear_positions()
```

```python
def find_gearbox_around(pos=(0,2), gear_pos=(1,3)):
    x,y = pos
    check_pos = []
    for xx in range(max(0, x-1), min(x+2, len(data))):
        line = []
        for yy in range(max(0, y-1), min(y+2, len(data[x]))):
            line.append((xx,yy))
        check_pos.append(line)


    for pos_line in check_pos:
        if gear_pos in pos_line:
            return True
        
    return False

find_gearbox_around()
```

```python
def extract_gear_box_number(
    gear_pos=(1,3),
    data=data
):
    gear_box_nums = []
    for line_index in range(max(0, gear_pos[0]-1), min(gear_pos[0]+2, len(data))):
        line = data[line_index]
        current_num = ""
        is_gearbox_around = False
        
        for index, c in enumerate(line):
            if c not in [f"{i}" for i in range(10)]:
                if is_gearbox_around:
                    is_gearbox_around = False
                    gear_box_nums.append(int(current_num))
                current_num = ""

            else:
                current_num = current_num + c
                is_gearbox_around = is_gearbox_around or find_gearbox_around((line_index, index), gear_pos)

        if is_gearbox_around:
            gear_box_nums.append(int(current_num))
            
    print(f"gear_box_nums = {gear_box_nums}")
    return gear_box_nums


extract_gear_box_number()
```
